```python
print("Average Marks")
print(data[['G1','G2','G3']].mean())
```

# **BAR CHART**

```python
plt.figure(figsize=(10,5))
data[['G1','G2','G3']].mean().plot(kind='bar')

plt.title("Average Student Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")

plt.show()
```

* G1, G2, and G3 average marks are similar
* Students performed slightly better in G2
* Overall performance is consistent

# **SCATTER PLOT**

```python
plt.figure(figsize=(8,5))
plt.scatter(data['G1'], data['G2'])

plt.xlabel("G1 Marks")
plt.ylabel("G2 Marks")
plt.title("G1 vs G2")

plt.show()
```



*   Strong relationship between    G1    and G2
* Students with high G1 marks also scored high in G2
* Few students show variation



# **HEATMAP**

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(numeric_only=True), annot=True)

plt.title("Correlation Heatmap")
plt.show()
```

* G1, G2, and G3 have strong correlation
* Study time impacts student marks
* Failures negatively affect marks

# **Conclusion**

* Students show consistent academic performance
* Previous exam marks influence final marks
* Study time improves performance
